```python
import pandas as pd

df = pd.read_csv("../data/raw/india-districts-census-2011.csv")

df.head()
```

```python
df.columns
```

```python
df.shape
```

```python
important_columns = [
    "District name",
    "State name",
    "Population",
    "Male",
    "Female",
    "Literate"
]

df = df[important_columns]

df.head()
```

```python
df.shape
```

```python
df.columns = [
    "district",
    "state",
    "population",
    "male_population",
    "female_population",
    "literate_population"
]

df.head()
```

```python
df["literacy_rate"] = (
    df["literate_population"] / df["population"]
) * 100
```

```python
df.head()
```

```python
df["district"] = df["district"].str.lower().str.strip()
df["state"] = df["state"].str.lower().str.strip()
```

```python
df.to_csv("../data/processed/district_master_dataset.csv", index=False)
```

```python
df["gender_ratio"] = (df["female_population"] / df["male_population"]) * 1000
```

```python
df["literacy_index"] = df["literacy_rate"] / 100
```

```python
df["population_weight"] = df["population"] / df["population"].max()
```

```python
df.head()
```

```python
df.to_csv("../data/processed/district_master_dataset.csv", index=False)
```
